# Labelling The Dataset

```python
temp = driver_standings.loc[driver_standings['Year']==1960].groupby(['Year','Car']).agg({'PTS':'sum'}).reset_index()
temp
```

```python
final_df = pd.DataFrame(columns=['Year','Car','PTS'])
driver_standings['Champion'] = 4
for i in list(driver_standings.Year.unique())[:-1]:
    temp = driver_standings.loc[driver_standings['Year']==i].groupby(['Year','Car']).agg({'PTS':'sum'}).reset_index() 
    # print(temp)
    pts = list(temp['PTS'].nlargest(3))
    max_pts = pts[0]
    sec_pts = pts[1]
    thir_pts = pts[2]
    temp['Champion'] = 4
    temp.loc[temp.PTS == max_pts,'Champion'] = 1
    temp.loc[temp.PTS == sec_pts,'Champion'] = 2
    temp.loc[temp.PTS == thir_pts,'Champion'] = 3
    final_df = final_df.append(temp)
    
final_df['Champion'] = final_df['Champion'].astype(int)
final_df.head()
```

```python
final_df.tail()
```

# Feature 1 : Average_points_Scored

```python
# Avweerage Race Points Scored each race
temp = race_details.groupby(['Year','Car']).agg({'PTS':'mean'}).reset_index()
temp.columns = ['Year','Car','Average_points_Scored']
final_df = final_df.merge(temp,left_on=['Year','Car'],right_on=['Year','Car'],how='left')
```

# Feature 2 : No_of_times_secured_front_grid

```python
# No of times they have secured front grid in Qualifying
qualifyings.head()
```

```python
def change(x):
    try:
        x = int(x)
        return x
    except:
        return 0
    
qualifyings['Pos'] = qualifyings['Pos'].apply(lambda x: change(x))
temp = qualifyings.groupby(['Year','Car','Pos']).agg({'Driver':'count'}).reset_index()
temp = temp.loc[temp.Pos.isin([1,2])]
temp.drop(['Pos'],axis=1,inplace=True)
```

```python
temp.columns = ['Year','Car','No_of_times_secured_front_grid']
temp = temp.groupby(['Year','Car']).agg({'No_of_times_secured_front_grid':'sum'}).reset_index()
temp.head()
```

```python
temp.tail()
```

```python
final_df = final_df.merge(temp,left_on=['Year','Car'],right_on=['Year','Car'],how='left')
final_df
```

```python
final_df['No_of_times_secured_front_grid'].fillna(0,inplace=True)
final_df.head()
```

# Feature 4 : Avg No of race wins each year

```python
final_df.head()
```

```python
race_summaries.head()
```

```python
temp = race_summaries.groupby(['Year','Car']).agg({'WinnerCode':'count'}).reset_index()
temp['Total_races'] = 0
for i in list(temp.Year.unique())[:-1]:
    temp1 = race_details.loc[race_details.Year == i]
    temp.loc[temp['Year']==i,'Total_races'] = temp1['Grand Prix'].nunique()

temp.head()
```

```python
temp['Win_percentage'] = temp['WinnerCode']/temp['Total_races']
temp.drop(['WinnerCode','Total_races'],axis=1,inplace=True)
temp.head()
```

```python
final_df = final_df.merge(temp,left_on=['Year','Car'],right_on=['Year','Car'],how='left')
final_df['Win_percentage'].fillna(0,inplace=True)
final_df.head()
```
